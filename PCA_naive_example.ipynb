```
import numpy as np

x1 = np.array([2.5, 0.5, 2.2, 1.9, 3.1, 2.3, 2, 1, 1.5, 1.1])
x2 = np.array([2.4, 0.7, 2.9, 2.2, 3.0, 2.7, 1.6, 1.1, 1.6, 0.9])

x2
```

```
mean_x1=np.mean(x1)
mean_x1
x1_norm = x1-mean_x1

mean_x2=np.mean(x2)
x2_norm = x2-mean_x2
print(x2_norm)

B=[x1_norm,x2_norm]
B=np.array(B)
print(B)
print(B.shape)
```

```
import matplotlib.pyplot as plt
fig = plt.figure()
plt.plot(x1_norm,x2_norm, '+', Color='r')
plt.grid()
plt.xlim((-2, 2))
plt.ylim((-2,2))
```

```
C=B@(B.transpose())/9
print(C)
sigma, vect = np.linalg.eig(C)

vect
```

```
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt


y1 = np.array([2.5, 0.5, 2.2, 1.9, 3.1, 2.3, 2, 1, 1.5, 1.1])
y2 = np.array([2.4, 0.7, 2.9, 2.2, 3.0, 2.7, 1.6, 1.1, 1.6, 0.9])
y3 = np.array([1.6, 1.7, 1.9, 1.8, 1.5, 1.7, 1.6, 1.8, 1.6, 1.9])

mean_y1=np.mean(y1)
y1_norm = y1-mean_y1

mean_y2=np.mean(y2)
y2_norm = y2-mean_y2

mean_y3=np.mean(y3)
y3_norm = y3-mean_y3

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(y1_norm,y2_norm, y3_norm, '+', Color='r')
plt.xlim((-2, 2))
plt.ylim((-2,2))
ax.set_zlim(-2, 2)
```

```
B1=[y1_norm,y2_norm, y3_norm]
B1=np.array(B1)
C1=B1@(B1.transpose())/9
print(C1)
sigma, vect = np.linalg.eig(C1)

print('sigma =', sigma)
print('eigenvectors =',vect)
```
